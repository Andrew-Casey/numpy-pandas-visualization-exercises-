#### 2. Change all the column names
      - lowercase letters
      - remove whitespace
      - remove periods

```python
#using a lambda function
df = df.rename(columns=lambda x: x.lower().replace(' ', '').replace('.', ''))
```

```python
#using list comprehension
df.columns = [col.lower().replace('.','').replace(' ','').strip() for col in df.columns]





```

```python
# using a for loop
new_columns = []
for col in df.columns:
    new_col = col.lower().replace(' ', '').replace('.', '')
    new_columns.append(new_col)
    
df.columns = new_columns
df
```

#### 6. Number of Pokemon whose Type 2 is Ghost

```python
type2_counts = df['type2'].value_counts()

# select the count for "Ghost"
num_ghost_type2 = type2_counts.loc['Ghost']

# display the result
print("Number of Pokemon whose Type 2 is Ghost:", num_ghost_type2)
```

#### 7. Percentage of Pokemon whose Type 2 is Ghost

```python
percentage_of_ghosts = round((num_ghost_type2/len(df)) *100,2)
percentage_of_ghosts
```

```python
# create a Boolean mask to filter the DataFrame to only include rows where "Type 2" is "Ghost"
mask = df['type2'] == 'Ghost'

# count the number of rows that match the mask
num_ghost_type_2 = mask.sum()

# calculate the percentage of Pokemon whose Type 2 is Ghost
percentage_ghost_type_2 = num_ghost_type_2 / len(df) * 100
percentage_ghost_type_2
```

#### 9. What is the slowest speed for Grass type (either type 1 or type 2)

```python
grass_mask = (df['type1'] == 'Grass') | (df['type2'] == 'Grass')
grass_pokemon = df[grass_mask]

# find the slowest speed for Grass type Pokemon
slowest_speed = grass_pokemon['speed'].min()
slowest_speed
```

```python
df[(df['type1'] == 'Grass') | (df['type2'] == 'Grass')].min()
```

#### 10. Find the average speed by Generation

```python
# group the DataFrame by "Generation" and calculate the mean of "Speed" for each group
avg_speed_by_gen = round(df.groupby('generation')['speed'].mean(),2)

# display the result
print("Average speed by Generation:")
print(avg_speed_by_gen)
```

# 11. How many LEGENDARY Pokemon are DRAGON type?

```python
mask = (df['type1'] == 'Dragon') & (df['legendary'] == True)
legendary_dragon_pokemon = df[mask]

# count the number of Legendary Dragon type Pokemon
num_legendary_dragon_pokemon = legendary_dragon_pokemon['legendary'].sum()

# display the result
print("Number of Legendary Dragon type Pokemon:", num_legendary_dragon_pokemon)
```
