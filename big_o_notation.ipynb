```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import factorial
# Define the x range
x = np.linspace(1, 10, 50)

# Define the y values for each Big O Notation curve
y1 = 0*x + 1
y2 = np.log(x)
y3 = x
y4 = x*np.log(x)
y5 = x**2
y6 = 2**x
y7 = factorial(x)
y8 = x**x

# Create a figure
fig, ax = plt.subplots(figsize=(8, 6))

# Add each Big O Notation curve to the plot
ax.plot(x, y1, label=r'$O(1)$')
ax.plot(x, y2, label=r'$O(\log n)$')
ax.plot(x, y3, label=r'$O(n)$')
ax.plot(x, y4, label=r'$O(n \log n)$')
ax.plot(x, y5, label=r'$O(n^2)$')
ax.plot(x, y6, label=r'$O(2^n)$')
ax.plot(x, y7, label=r'$O(n!)$')
ax.plot(x, y8, label=r'$O(n^n)$')

# Add a title to the figure
ax.set_title('Big O Notation')

# Label the x and y axes
ax.set_xlabel('Elements')
ax.set_ylabel('Operations')

# Add a legend to the figure
ax.legend(loc='best')

# Show the plot
plt.show()
```
