# 1

Use matplotlib to plot the following equation:

y = x**2 - x + 2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x range
x = np.linspace(-10, 10, 100)

# Calculate the corresponding y values for the given equation
y = x**2 - x + 2

# Plot the curve
plt.plot(x, y)

# Annotate the point (0,0)
plt.annotate('(0,0)', xy=(0, 0), xytext=(1, 10),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Set the axis labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('y = x**2 - x + 2')

# Show the plot
plt.show()
```

# 3

Combine the figures you created in the last step into one large figure with 4 subplots

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the x ranges
x1 = np.linspace(0, 10, 100)
x2 = np.linspace(-5, 5, 100)
x3 = np.linspace(-5, 5, 100)
x4 = np.linspace(-10, 10, 100)

# Calculate the corresponding y values for each equation
y1 = np.sqrt(x1)
y2 = x2**3
y3 = 2**x3
y4 = 1/(x4+1)

# Create a figure with 4 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Add the first subplot for y = sqrt(x)
axs[0, 0].plot(x1, y1)
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')
axs[0, 0].set_title('y = sqrt(x)')

# Add the second subplot for y = x**3
axs[0, 1].plot(x2, y2)
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')
axs[0, 1].set_title('y = x**3')

# Add the third subplot for y = 2**x
axs[1, 0].plot(x3, y3)
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')
axs[1, 0].set_title('y = 2**x')

# Add the fourth subplot for y = 1/(x+2)
axs[1, 1].plot(x4, y4)
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Y')
axs[1, 1].set_title('y = 1/(x+1)')

# Add a title for the entire figure
fig.suptitle('Four Different Equations')

# Adjust the layout to avoid overlapping labels
plt.tight_layout()

# Show the plot
plt.show()
```

# 4

Combine the figures you created in the last step into one figure where each of the 4 
equations has a different color for the points. Be sure to include a legend and an 
appropriate title for the figure.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the x ranges
x1 = np.linspace(0, 10, 100)
x2 = np.linspace(-5, 5, 100)
x3 = np.linspace(-5, 5, 100)
x4 = np.linspace(-10, 10, 100)

# Calculate the corresponding y values for each equation
y1 = np.sqrt(x1)
y2 = x2**3
y3 = 2**x3
y4 = 1/(x4+2)

# Create a figure
fig, ax = plt.subplots(figsize=(10, 8))

# Add the first set of points for y = sqrt(x)
ax.plot(x1, y1, label='y = sqrt(x)', color='blue')

# Add the second set of points for y = x**3
ax.plot(x2, y2, label='y = x**3', color='green')

# Add the third set of points for y = 2**x
ax.plot(x3, y3, label='y = 2**x', color='red')

# Add the fourth set of points for y = 1/(x+2)
ax.plot(x4, y4, label='y = 1/(x+2)', color='purple')

# Add a title for the figure
ax.set_title('Four Different Equations with Different Colors')

# Add a legend to the figure
ax.legend(loc='best')

# Show the plot
plt.show()
```
