Plot the x and y values from the anscombe data. Each dataset should be in a separate column.

```python
# Plot x and y values for each dataset
fig, axs = plt.subplots(1, 4, figsize=(12, 4), sharex=True, sharey=True)

for i, (dataset, group) in enumerate(anscombe.groupby('dataset')):
    ax = axs[i]
    ax.scatter(x=group['x'], y=group['y'])
    ax.set_title(f"Dataset {dataset}")
    
plt.show()
```

Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
from env import host_name, user, password
```

```python
def get_db_url(database):
    url = f'mysql+pymysql://{user}:{password}@{host_name}/{database}'
    return url
```

```python
url = get_db_url('chipotle')
```

```python
orders = pd.read_sql('SELECT quantity, item_name, item_price FROM orders', url)
orders
```

```python
orders['item_price'] = orders['item_price'].str.replace('$', '', regex=False)
orders
```

```python
# group the data by item name and calculate the revenue for each item
orders.item_price = orders.item_price.astype(str).str.replace('\$', '', regex=True).astype(float)
orders
```

```python
# Get the data for the revenue generated by the top 4 items
revenue_generated = orders.groupby('item_name')['item_price'].sum().sort_values(ascending=False).head(5)
revenue_generated
```

```python
# sort the items by revenue in descending order and select the top 4 items
pop_items = orders.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5)
pop_items
```

```python
# Get the data for the revenue generated by the top 4 items
revenue_generated = orders.groupby('item_name')['item_price'].sum().sort_values(ascending=False).head(5)

# Get the data for the top 4 most popular items by quantity sold
pop_items = orders.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5)

# Merge the two dataframes
merged_items = pd.merge(pop_items, revenue_generated, on='item_name')

# Set the plot title and axis labels
sns.barplot(x=merged_items.index, y=merged_items['quantity'], label='Quantity')
sns.barplot(x=merged_items.index, y=merged_items['item_price'], label='Revenue')
plt.title('Top 4 Items by Quantity Sold and Revenue Generated')
plt.xlabel('Item Name')
plt.ylabel('Quantity Sold / Revenue ($)')
# Add a legend
plt.legend()

# Show the plot
plt.show()
```

Load the sleepstudy dataset from pydataset and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.

```python
# Load the sleepstudy dataset
sleepstudy = data('sleepstudy')

# Create line chart of all the individual subject's reaction times
sns.set_style("whitegrid")
sns.set_palette('tab10')
sns.set(rc={'figure.figsize':(16,8)})
sns.lineplot(x='Days', y='Reaction', data=sleepstudy, hue='Subject', alpha=0.7)

# Create more prominent line showing the average change in reaction time
sns.lineplot(x='Days', y='Reaction', data=sleepstudy, color='black', label='Average')

# Set the plot title and axis labels
plt.title('Sleep Study Reaction Times')
plt.xlabel('Days')
plt.ylabel('Reaction Time')

# Show the plot
plt.show()
```
